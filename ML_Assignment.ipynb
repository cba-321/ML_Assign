Question 1

Matrix multiplication is the binary operation. The resultant matrix,say C, of the multiplication of the two matrices, lets say A, and B, is the dot product of the corresponding row of matrix A and column of matrix B.

if we have a matrix A (n,k), and matrix B (k,m), then the dimension of the matric C will be (n,m) and the specific element of the of the C[i,j] would be:

C[i,j] = A[i,1] * B[1,j] + A[i,2] * B[2,j] + A[i,3] * B[3,j] +.....+ A[i,k-1] * B[k-1,j] + A[i,k] * B[k,j].

It means to calculate a single element of the matrix C, we need to perform 'k' multiplication and 'k-1' addition. Since there are 'n' rows in matrix C, so to find the element of a row, we need to perform 'n * k' multiplication, and 'n * (k-1)' addtion. Since, there are m columns in matrix C, to calculate all elements of C, we need to perform 'n * m * k'multiplictions, and 'n * m * k-1' additions.

```python
import numpy as np
import timeit
```

```python
## matrix multiplication using list
def using_list(matrix_A, matrix_B):
    result = [[0 for _ in range(len(matrix_B[0]))] for _ in range(len(matrix_A))]
    
    for a in range(len(matrix_A)):
        for b in range(len(matrix_B[0])):
            for c in range(len(matrix_B)):
                result[a][b] += matrix_A[a][c] * matrix_B[c][b]
    
    return result

# Example matrices
matrix_A = [[1, 2], [3, 4], [5, 6]]
matrix_B = [[7, 8], [9, 10]]

## Calculating time using list
Time_list = %timeit -o using_list(matrix_A, matrix_B)

## matrix multiplication using numpy
np_matrix_A = np.array(matrix_A)
np_matrix_B = np.array(matrix_B)

resultant = np.dot(np_matrix_A, np_matrix_B)

Time_numpy = %timeit -o resultant
print(f"Time using lists: {Time_list.best:.10f} seconds")
print(f"Time using NumPy: {Time_numpy.best:.10f} seconds")
```

Time using Numpy is less than the time using lists. This is because numpy uses vectorised operation, means it do all calculation on the array at once while list nested loop to do the calculation. Also numpy is written in c, which is low-level language and faster than the interpreted code used by list.

```python
import numpy as np
import pandas as pd

#create matrix of 20x5 ranging from 1 to 2 using numpy
matrix = np.random.uniform(1, 2, (20,5))

#Create a dataframe using panda
df = pd.DataFrame(matrix, columns = ['a', 'b', 'c', 'd', 'e'])

#find column with highest standard deviation
#max() give the maximum value while idxmax() give the index of the maximum value
MaxStd_col = df.std().idxmax()

#find row with lowest mean

MinMean_row = df.mean(axis = 1).idxmin()

print("DataFrame")
print(df)
print("\n column with highest standard deviation:", MaxStd_col)
print("\n row with lowest mean:", MinMean_row)
```

```python
df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

df['g'] = df['f'].apply(lambda x: 'LT8' if x<8 else 'GT8')

#number of rows with column g's value GT8
num_row = (df['g'] == 'LT8').sum()


std_LT8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_GT8 = df.loc[df['g'] == 'GT8', 'f'].std()

print(df)
print(num_row)
print(std_LT8)
print(std_GT8)
```

Broadcasting is a method which allows us to do airthmetic operation on arrays of different sizes. The shorter array broadcasted to the large array so that element-wise operation can be performed on them.

```python
import numpy as np

#let's create two 2D arrays of different sizes

A = np.array([[3,6,1], [1,5,3]])
B = np.array([2, 4, 1])

# B is broadcasted to match the shape of A
result = A + B

print("result of broadcasting:\n", result)
```

```python
import numpy as np

def argmin(arr):
    
    if len(arr) == 0:
        raise ValueError("Input array is empty")

    # Initialize variables to store the minimum value and its index
    min_value = arr[0]
    min_index = 0

    # Iterate through the array to find the minimum value and its index
    for i, value in enumerate(arr):
        if value < min_value:
            min_value = value
            min_index = i

    return min_index
```
